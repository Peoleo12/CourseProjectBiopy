```python
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
#from Bio.Alphabet import generic_dna
from Bio import SeqIO
import random
```

```python
num_seqs = 2
seq_length = 50

records = []
for i in range(num_seqs):
    seq = ''.join(random.choices(['A', 'C', 'G', 'T'], k=seq_length))
    # New styl
    record = SeqRecord(Seq(seq), id=f"Seq{i}", description="random DNA sequence")
    records.append(record)

SeqIO.write(records, "sequences.fasta", "fasta")
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
x='TAACTGATTGCCCACCCATTTAGGCAACCATTATGCCATCCGGGGCCGCC'
y='CGGATGTGTATCGCTTGGAGCCTTGAAAACCAACAGTTCCGAGAAAGGGC'
alignments = pairwise2.align.localxx(x, y)
for a in alignments:
 print(format_alignment(*a))
```

```python
from Bio.pairwise2 import format_alignment

x = "TAACTGATTGCCCACCCATTTAGGCAACCATTATGCCATCCGGGGCCGCC"
y = "CGGATGTGTATCGCTTGGAGCCTTGAAAACCAACAGTTCCGAGAAAGGGC"

match_score = 2
mismatch_penalty = -1
gap_open_penalty = -0.5
gap_extend_penalty = -0.1

alignments = pairwise2.align.globalms(x, y, match_score, mismatch_penalty, gap_open_penalty, gap_extend_penalty)

for a in alignments:
  print(format_alignment(*a))
```

```python
from Bio.Blast import NCBIWWW, NCBIXML

# Run a BLAST search for a given query sequence
result_handle = NCBIWWW.qblast("blastn", "nr", "TAACTGATTGCCCACCCATTTAGGCAACCATTATGCCATCCGGGGCCGCC")

# Parse the search results
blast_record = NCBIXML.read(result_handle)

# Do something with the results, such as print out the top 10 hits
for alignment in blast_record.alignments[:10]:
    print(alignment.title)
```
